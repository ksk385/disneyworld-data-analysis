```python
import matplotlib.pyplot as plt
df.loc[df['Activity Type'] == 'Attraction', 'Height Restrictions'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.title('Attractions Height Restrictions (inches)')
plt.show()
```
